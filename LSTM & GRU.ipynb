```python
import pandas as pd

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df = pd.read_csv(url)

df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(df['Temp'])
plt.title("Daily Minimum Temperature")
plt.show()
```

```python
import numpy as np

data = df['Temp'].values.astype(float)
seq_len = 30

X, y = [], []

for i in range(len(data) - seq_len):
    X.append(data[i:i+seq_len])
    y.append(data[i+seq_len])

X = np.array(X)
y = np.array(y)

X.shape, y.shape
```

```python
X = X.reshape((X.shape[0], X.shape[1], 1))
X.shape
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential()

# üî• LSTM layer
# 50 units = ÿ≠ÿ¨ŸÖ ÿßŸÑÿ∞ÿßŸÉÿ±ÿ©
# return_sequences=False ‚Üí ÿ¢ÿÆÿ± hidden state ŸÅŸÇÿ∑
model.add(layers.LSTM(50, input_shape=(30, 1)))

# üî• Linear output layer
model.add(layers.Dense(1))

model.compile(
    optimizer="adam",
    loss="mse"
)

model.summary()
```

```python
history = model.fit(
    X, y,
    epochs=10,
    batch_size=32,
    verbose=1
)
```

```python
test_seq = X[-1].reshape(1, 30, 1)
pred = model.predict(test_seq)
pred
```

```python
final_loss = history.history['loss'][-1]
print("Final Training Loss:", final_loss)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.title('Training Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.show()
```

# **GRU**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models

# Load dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df = pd.read_csv(url)

data = df['Temp'].values.astype(float)
seq_len = 30

# Prepare sequences
X, y = [], []
for i in range(len(data) - seq_len):
    X.append(data[i:i+seq_len])
    y.append(data[i+seq_len])

X = np.array(X)
y = np.array(y)

# Reshape for GRU
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
model = models.Sequential()

# GRU layer
model.add(layers.GRU(50, input_shape=(seq_len, 1)))

# Output layer
model.add(layers.Dense(1))

# Compile
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
history = model.fit(
    X, y,
    epochs=20,
    batch_size=32,
    validation_split=0.1,  # 10% for validation
    verbose=1
)
```

```python
plt.figure(figsize=(8,4))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('GRU Training and Validation Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.show()
```
